```python
df = odf[['site_no', 'texture', 'year', 'material_id', 'specimen_status', 'city', 'state',
            'internal_drainage', 'resistivity', 'pH', 'mean_temp', 'annual_precipitation', 'moisture',
            'iron_type', 'areation','approximate_days',
            'weight_loss_g', 'corrosion_rate', 'C_clr',
            'Si_clr', 'Mn_clr', 'S_clr', 'P_clr', 'Cr_clr', 'Ni_clr', 'Cu_clr',
            'Mo_clr', 'C (%)', 'Si (%)', 'Mn (%)', 'S (%)',
            'P (%)', 'Cr (%)', 'Ni (%)', 'Cu (%)', 'Mo  (%)', 'Fe (%)',
       ]].copy()
```

```python
df['log_days'] = np.log(df['approximate_days'])
```

```python
df['moisture_res_index'] = df['moisture'] / df['resistivity']
```

```python
df['T_over_D'] = df['thickness'] / df['nominal_width_diameter']
```

```python
def climate_zone(row):
    if row['mean_temp'] > 60 and row['annual_precipitation'] > 40:
        return 'Hot‑Humid'
    elif row['mean_temp'] < 45 and row['annual_precipitation'] < 20:
        return 'Cold‑Dry'
    else:
        return 'Temperate'
df['climate_zone'] = df.apply(climate_zone, axis=1)
```

```python
df['log_resistivity']   = np.log(df['resistivity'])
df['log_corrosion_rate'] = np.log(df['corrosion_rate'])
```

```python
rows_with_na = df[df.isna().any(axis=1)]
```

```python
df = pd.get_dummies(df, columns=['texture'], prefix='texture', drop_first=True, dtype=int)  
```

```python
df = pd.get_dummies(df, columns=['iron_type'], prefix='iron_type', drop_first=True, dtype=int) 
```

```python
df = pd.get_dummies(df, columns=['material_id'], prefix='material_id', drop_first=True, dtype=int)
```

```python
df = pd.get_dummies(df, columns=['form'], prefix='form', drop_first=True, dtype=int) 
```

```python
df = pd.get_dummies(df, columns=['climate_zone'], prefix='climate_zone', drop_first=True, dtype=int) 
```

```python
order = {'G': 3,    # Good   → 3 (best)
         'F': 2,    # Fair   → 2
         'P': 1,    # Poor   → 1
         'VP': 0}   # Very Poor → 0 (worst)

df['internal_drainage_code'] = df['internal_drainage'].map(order)
```

```python
aeration_order = {
    "Good": 3,
    "Fair": 2,
    "Poor": 1,
    "Very Poor": 0
}

df["areation_code"] = df["areation"].map(aeration_order)
```
